```python
fac_df = pd.DataFrame()

# try to factorize the columns so we can find which are redundant
for col in cols:
    try:
        fact_col = df[col].factorize()
        fac_df[col] = fact_col[0]
    except:
        pass
```

```python
corr_df = fac_df.corr()
count = 0
for col1 in corr_df.columns:
    for col2 in corr_df.columns:
        if col1 != col2 and corr_df[col1][col2] == 1:
            print("Col:", col1, "Col:", col2, corr_df[col1][col2])
            count += 1
            
print("Num correlated cols:", count)
```
