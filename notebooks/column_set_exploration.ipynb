```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = np.load("columns.npy")
df = pd.read_pickle("data/df.pkl")
```

```python
# subset to only the 71 columns we will be using
df = df[sorted(cols)]
```

```python
df.head()
```

## Pie Charts of Values in Selected Columns

```python
# see values for columns which have a small set of possible values
for col in df.columns:
    try:
        counts = pd.value_counts(df[col], normalize=True)
        if len(counts) < 25:
            plt.pie(counts, labels=counts.index, shadow=True)
            plt.title(col)
            plt.show()
#             print("Column:", col)
#             print(counts)
#             print()
    except:
        pass
```
